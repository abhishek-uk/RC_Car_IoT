# RC car: IoT

### This is a documentation for RC car that have been controlled by Mobile 

Components Required,
<br>    1. Car Frame
<br>    2. Motor for rear wheel (I use a 6v dc motor)
<br>    3. Servo motor SG90 for Steering
<br>    4. nodeMCU 
<br>    5. Single H bridge motor driver 
<br>    6. Battery (5v recommended, max 10v)
<br>    7. Bunch of cables

## Step 1

First select a Car frame that we can work and can store all of our components

Little modification on the frame for perfectly align our components 

Place rear motor with rear wheel. Better use a gear with it.

Next design a Steering mechanism for front wheel 

Attach the servo motor to the steering mechanism
